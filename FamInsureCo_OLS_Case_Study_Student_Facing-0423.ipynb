### Step 6: <span style="color:teal"> Adding legends to your Scatter Plot Visualization </span>


<b> Having created the respective scatter plot visualizations, color each data point to indicate the respective `bmi_classification` and add this legend at the bottom of the 3 scatter plot sub plots. </b>

<h3>... But how can we tell which scatter plot refers to which weight classification? </h3>

This is where we'll need to make use of the `c` argument in our `.plot()` method, which is shorthand for color.
In a nutshell, we're going to have to define a <b> dictionary object </b> that maps the `bmi_classification` to a specific color.

In other words, something like this:

`colors = {'Underweight': 'Orange', 'Healthy': 'green', 'Slightly Overweight': 'blue', 'Overweight': 'red' }`

For a list of colors, go <a href = "https://matplotlib.org/stable/gallery/color/named_colors.html"> here</a>

We would then <b> map </b> this back to our plot accordingly using the `c` argument.

`df.plot(kind='scatter', x='a', y='b', c=df.index.map(colors))`

Think of the `.map()` as applying the color dictionary to the `df['bmi_classification']` column, which is now the `index` due to the `group by`.


Lastly, with the colors now applied, we need to create the legend.

#### 1. Specify the 3 sub plots plotting coordinates
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

#### 2. Specify the color dictionary for our scatter plot
    colors = {'Underweight': 'Orange', 'Healthy': 'green', 'Slightly Overweight': 'blue', 'Overweight': 'red' }

#### 3. Plotting the scatter plots with color categories
    df.plot(kind='scatter', x='count', y='mean', c=df.index.map(colors), ax=axes[0])
    axes[0].set_title("Count vs Mean")
    
    Repeat this for the 2nd and 3rd Scatter Plots

#### 4. Lastly, add the legend in for the respective color scales.
    handles = []
    for label, color in colors.items():
        handles.append(plt.Line2D([], [], marker='o', color=color, label=label, linestyle='None'))
    fig.legend(handles, colors.keys(), loc='center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=4)

    plt.tight_layout()
    plt.show()


Now it's your turn to implement this below!

<h2 style="color:teal"> Write your answer below </h2>

````python
colors = {'Underweight': 'Orange', 'Healthy': 'green', 'Slightly Overweight': 'blue', 'Overweight': 'red' }
fig, axes = plt.subplots(1,3, figsize=(15,5))

agg_df.plot(kind='scatter', x='count', y='mean', c=agg_df.index.map(colors), ax=axes[0])
axes[0].set_title("Count vs Mean")
agg_df.plot(kind='scatter', x='count', y='sum', c=agg_df.index.map(colors), ax=axes[1])
axes[1].set_title("Count vs Sum")
agg_df.plot(kind='scatter', x='count', y='std', c=agg_df.index.map(colors), ax=axes[2])
axes[2].set_title("Count vs Std")

handles = []
for label, color in colors.items():
    handles.append(plt.Line2D([], [], marker='o', color=color, label=label, linestyle='None'))
fig.legend(handles, colors.keys(), loc='center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=4)

plt.tight_layout()
plt.show()
````

### Step 12: <span style="color:teal"> Create Quartiles</span>

The general logic for removing an outliers is:

1. Define the Quartiles Q1 and Q3.
2. Calculate the Interquartile Range (Q3 - Q1)
3. Create two new variables; Lower_Range (Q1 - 1.5x IQR) and Upper_Range ( Q3 + 1.5x IQR).

### Step 13: <span style="color:teal"> Identify & Remove Outliers</span>

Continuing in from Step 12 - we still need to calculate <b>how</b> we can remove our outliers.

We're aware of the overall formula as shown per below.

<b> Outlier = Q1 - 1.5x IQR OR Q3 + 1.5x IQR </b>

Now let's work out how we can identify these Outliers step by step.

<b> i) Define two new variables, Lower_Limit and Upper_Limit can be calculated as:</b>
<p>
<li> Lower_Limit = Q1 - 1.5 * IQR </li>
<li> Upper_Limit = Q3 + 1.5 * IQR </li>
   
<b> ii) Using Lower_Limit and Upper_Limit with an OR (|) condition, filter the dataframe to include *only* the Outliers. How many Outliers do we have?</b>
    
<b>Now note that we only want to use the *upper limit* for our filtering.
This is because the lower limit has us retrieve -ve medical insurance charges, which isn't valid. </b>

What this means is that the syntax should look like this:
    
    intermediate_df_outliers = some_dataframe [ (some_dataframe > Upper_Limit)]
    
If you're stuck, <a href = https://stackoverflow.com/questions/54398554/how-to-remove-outliers-in-python> here's </a> a link that will show you how you can remove Outliers in Python!

<b> iii) What is the percentage of data that remains after we've removed the Outliers from the `FamInsureCo` dataset?</b>
    

<h2 style="color:teal"> Write your answer below </h2>

````python
Lower_Limit = Q1 - 1.5 * IQR
Upper_Limit = Q3 + 1.5 * IQR

#intermediate_df_outliers = intermediate_df[(intermediate_df > Upper_Limit).any(axis=1)]
intermediate_df_outliers = intermediate_df[intermediate_df['medical_insurance_charges'] > Upper_Limit]






percentage_remaining = (len(intermediate_df) - len(intermediate_df_outliers)) / len(intermediate_df) * 100

print(f"Percentage of data remaining after outlier removal: {percentage_remaining:.2f}%")
````

## Appendix: Python Cheat Sheet

**SELECTING COLUMNS**

```Python
- dataframe['a'] -- Select column a from a dataframe individually
- dataframe[['a','b']] -- Select columns a and b from a dataframe
- dataframe[:5] -- Select the first five rows from the dataframe
- dataframe[2:] -- Select all the rows of the dataframe starting from row 2 onwards
- dataframe.head(n) -- Select the number of rows in the dataframe specified by n
- dataframe.loc[row,column] -- Selects the number of rows, and columns specified with labels
- dataframe.iloc[row,column_index] -- Select the number of rows, and columns referencing their index

```

**LIBRARIES**
```Python
- import pandas as pd -- This imports the pandas library that contains the methods you need to manipulate tabular data
```

**OPERATORS**
```Python

- > - Greater than
```

**DESCRIPTIVE STATISTICS FUNCTIONS**

**SORTING VALUES**
